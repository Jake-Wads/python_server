## Probability Distributions

A scaling of possible event outcomes based upon their likelihood (probability) of occurring and described by a probability function.  

### 1. Discrete probability distribution
A probability distribution where each class contains only certain values of the variable in any particular interval (such as only whole number values, for example).  

#### a. Binomial Distribution:  
- The number of successes in a sequence of n independent experiments.     
- Examples: yes/no, true/false, present/absent, on/off, heads/tails.   
- $p \in [0,1]$

#### c. Poisson Distribution:
- Given mutually independent events, this distrbution models that number of times an event occurs in an interval of time or space.    
- Assumptions:   
    - the occurrence of one event does not affect the probability that a second event will occur. (independence)   
    - the rate at which the events occur is constant.    
    - two events cannot occur at exactly the same instant    
- If the actual distribution is binomaial and the number of trials is sufficiently bigger than the number of successes, a Poisson may be appropriate.

#### c. Normal Distribution:
- a mean of $\mu$ and a standard deviation of $\sigma$
- If $X \thicksim N(\mu, \sigma^2)$ then $Z = \frac{X - \mu}{\sigma} \thicksim N(0,1)$   
- If $Z$ is standard normal then $X = \mu + \sigma Z \thicksim N(\mu,\sigma^2)$        
- Symmetrical probability distribution

### 3. Probability Distribution Skewness
1. Symmetric
2. Left-skewed: A set of data values in which the mean is generally less than the median. The left tail of the distribution is longer than the right tail of the distribution.  
3. Right-skewed: A set of data values in which the mean is generally greater than the median. The right tail of the distribution is longer than the left tail of the distribution.
