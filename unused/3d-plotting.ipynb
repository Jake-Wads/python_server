```python
import pandas as pd
import numpy as np
import pydataset

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

mtcars = pydataset.data('mtcars')
mtcars.head()
```

```python
fig = plt.figure()
ax = Axes3D(fig)

ax.scatter(mtcars.hp, mtcars.wt, mtcars.mpg, c=mtcars.cyl)
ax.set(xlabel='hp', ylabel='wt', zlabel='mpg')
```

```python
%matplotlib notebook

fig = plt.figure()
ax = Axes3D(fig)

x = np.arange(1, 51)
y = x + sorted(np.random.rand(50))
z = np.random.randn(50)

ax.plot(x, y, z)
ax.set(xlabel='x', ylabel='y', zlabel='z')
```

```python
fig, axs = plt.subplots(1, 2, subplot_kw={'projection': '3d'})
```

```python
%matplotlib notebook

fig = plt.figure()
ax = Axes3D(fig)

x = np.linspace(-8, 8, 100)
y = np.sin(x)
z = np.cos(x)

ax.plot(x, y, z)
ax.set(xlabel='x', ylabel='y', zlabel='z')
```

```python
%matplotlib notebook

fig = plt.figure()
ax = Axes3D(fig)

x = np.linspace(-4, 4, 100)
y = np.linspace(-4, 4, 100)
z = np.cos(x + y)

ax.plot(x, y, z)
ax.set(xlabel='x', ylabel='y', zlabel='z')
```
