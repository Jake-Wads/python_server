```python
import pandas as pd
```

```python
df1 = pd.read_csv("SA311_2018.csv")
df2 = pd.read_csv("SA311_2017.csv")
df = df1.append(df2)
```

```python
df = df.drop(columns=['CASEREF', 'CLIENTID', 'SUBJECTNAME', 'XCOORD','YCOORD',
                      ' interactiondate','ALLOCATEDTODEPTID', 'Report Starting Date',
                      'Report Ending Date', 'Select Dept?'])

df.columns.values
```

```python
df = df.rename(index=str, columns={'CASEID': 'case_id', 'OPENEDDATETIME': 'case_opened_date',
                                   'CLOSEDDATETIME': 'case_closed_date', 
                                   'SLA_Date': 'SLA_due_date',
                                   'Subject To SLA': 'dept_subject_to_SLA', 
                                   'Late (Yes/No)': 'case_late',
                                   'LATEDAYS': 'num_days_late', 'Closed (Yes/No)': 'case_closed',
                                   'Dept': 'dept_name', 'REASONNAME': 'dept_division',
                                   'TYPENAME': 'service_request_type',
                                   'SLADays': 'SLA_days', 'CaseStatus': 'case_status',
                                   'SOURCEID': 'source_id', 'SOURCEUSERID': 'source_username',
                                   'OBJECTDESC': 'request_address',
                                   'Council District': 'council_district',
                                   'Dept (to)': 'standardized_dept_name'
                             })
```

```python
source311 = df[['source_id','source_username']].drop_duplicates() \
            .sort_values(by=['source_id','source_username']) \
            .reset_index() \
            .drop(columns='index')

dept311 = df[['dept_division','dept_name','standardized_dept_name',
             'dept_subject_to_SLA']].drop_duplicates() \
            .sort_values(by=['dept_division','dept_name','standardized_dept_name']) \
            .reset_index() \
            .drop(columns='index')

case311 = df.drop(columns = ['source_username', 'dept_name',
                             'standardized_dept_name', 'dept_subject_to_SLA'])
```
