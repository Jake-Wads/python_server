# Compare Means

## Using the t-test to test for significant difference between means of 2 groups

## Set up the test

1. My question: Is the monthly invoice of the internet only customers in their 4th year significantly different than that of similar customers in their first year?  

2. State your hypothesis:   

    - $H_{0}$: The monthly invoice of the internet customers in their 4th year is no different from that of similarcustomers in their first year.  

    - $H_{1}$: The monthly invoice of the internet customers in their 4th year is different than that of similar customers in their first year.  

3. Prepare the data for the 2 groups you wish to compare

4. Is the distribution fairly normal?  Using the histogram and comparing the median=42 and mean=43, we can say the distribution is normal.  Therefore, we can feel comfortable using a t-test.  

5. Is it a one-tailed or two-tailed test? are you curious about any difference or a specific direction.  An example of a one-tailed would be:  Is the monthly invoice of internet customers in their 4th year higher than that of similar customers in their first year?   In our case, we are doing a two-tailed t-test

6. Set the desired confidence interval & significance level for your test.  Generally, this will be 95% CI => .05% $\alpha$.  

7. We have a set sample size, so we can't adjust that up.

    - Group 1 sample size = 159
    - Group 2 sample size = 87

8. Statistical Power = .80

9. Effect size:  we will not address calculating in excel, but we will when we get to python.

10. Compute the standard deviations for each group to see if we should assume equal variance.

Conclusion: We do a two-tail test (inequality). lf t Stat < -t Critical two-tail or t Stat > t Critical two-tail, we reject the null hypothesis. Is -4.78 < -1.96 OR -4.78 > 1.96? 

Because -4.78 < -1.96, we reject the null hypothesis. 

The observed difference between the sample means (56.1 - 61.4) is convincing enough to say that the average monthly invoice of customers in their 1st year vs. their 4th year differ significantly.

### Confidence Interval

- confidence = CONFIDENCE.T(alpha, standard_deviation, sample_size)
- upper bound = mean + confidence
- lower bound = mean - confidence
