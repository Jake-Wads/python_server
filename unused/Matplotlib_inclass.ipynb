```python
x = range(-3, 10)
y = [n * 2 for n in x]
z = [n / 2 for n in x]

plt.plot(x, y, c='orange')
plt.plot(x, z, c='green')

plt.annotate('Intersection', xy=(0,0), xytext=(-3, 5),
             arrowprops={'facecolor': 'blue'})

plt.show()
```

```python
x = [-5, -3, 0, 3, 5]
sin_x = [math.sin(n) for n in x]
cos_x = [math.cos(n) for n in x]

plt.figure(figsize=(14,6)) # width, height

print(x)
print(sin_x)
print(cos_x)
```

```python
plt.scatter(x, sin_x, c='blue')
plt.scatter(x, cos_x, c='orange')
plt.plot(x, sin_x, c='blue', ls = ':', label='$\sin(x)$')
plt.plot(x, cos_x, c='orange', ls = ':', label='$\cos(x)$')

plt.legend(loc='best')
plt.title('Approx sine and cosine curves')
plt.xlabel('$x$')

plt.show()
```

```python
x = [1, 1, 2, 3, 3, 3, 4, 4, 5]
bins = [0, 1, 2, 3, 4, 5, 6]
plt.figure(figsize=(16,6))
plt.suptitle(f'x = {x}')

plt.subplot(131)
plt.hist(x, bins=bins)
plt.title('align="middle" (default)')

plt.subplot(132)
plt.hist(x, bins=bins, align='left')
plt.title('align="left"')

plt.subplot(133)
plt.hist(x, bins=bins, align='right')
plt.title('align="right"')

plt.show()
```

```python
plt.figure(figsize=(12,8))

for i in range(1, 5):
    n = 10 ** i
    x1 = [randint(1, 6) for _ in range(n)]
    x2 = [randint(1, 6) for _ in range(n)]

    total = [x + y for x, y in zip(x1, x2)]

    plt.subplot(2, 2, i)
    plt.hist(total, bins=range(2, 14), align='left', color='white', edgecolor='black')
    plt.title(f'n = {n}')

plt.suptitle('Sum of 2 dice rolls')
plt.show()
```
