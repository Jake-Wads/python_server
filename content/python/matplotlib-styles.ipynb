# Matplotlib Styles


```python
from matplotlib import pyplot as plt
```

```python
plt.style.available
```

```python
def plot_with_style(style):
    import math

    x = range(-4, 4)
    y = [math.sin(n) for n in x]

    plt.style.use(style)
    plt.plot(x, y)
    plt.title(style)
    plt.scatter(x, y, s=100)
    plt.show()
```

```python
for style in plt.style.available:
    plot_with_style(style)
```
